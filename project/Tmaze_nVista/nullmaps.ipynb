```python
def place_cell_idx(task, name, day, cond):
    #detect place cells by spatial information.
    
    dfs = load_dfsdff(task, name, day)
    dfs = smoothing(dfs, sigma=3)
    dfs = dfs.query(cond)
    
    dfss = null_maps(dfs, 100)
    spinfo_o = spinfo(dfs, cond)
    spinfos = [spinfo(d,cond) for d in dfss]
    idx = [True if np.percentile([spi[n] for spi in spinfos], 95) < spinfo_o[n] else False for n in range(len(spinfo_o))]
    
    root_path = Path(r"F:\Data\Tmaze\Tmaze nVista\processed_data")
    cond2 = cond.replace(' ','').replace('>',"MT").replace('<',"LT") + ".txt"
    output_file = root_path/task/name/day/"place_cell_properties.csv"
    if not output_file.exists():
        p = pd.DataFrame([])
    else:
        p = pd.read_csv(output_file)
    p[cond] = idx
    p.to_csv(output_file, index=False)
```
