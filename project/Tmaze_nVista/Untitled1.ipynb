```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from utils import load_dfs, pr
```

```python
pd.set_option('display.max_rows', 70)
pd.set_option('display.max_columns', 70)
```

```python
task, name, day = "Tmaze_random", "A1", "d12"
```

```python
df = load_dfs(task, name, day, pr); df
```

```python
A_trials = df[df.type=='A'] ;AA_trials
B_trials = df[df.type=='B'] ;BB_trials
```

```python
A_spatial_maps = A_trials.groupby('bin').mean().loc[:,'cell0':]; A_spatial_maps
B_spatial_maps = B_trials.groupby('bin').mean().loc[:,'cell0':]; B_spatial_maps
```

```python
spatial_maps = pd.concat([A_spatial_maps, B_spatial_maps], axis=0); spatial_maps
```

```python
cor = spatial_maps.T.corr(); spatial_maps.T.corr()
```

```python
sns.heatmap(cor)
plt.ylabel("B                              A")
plt.xlabel("A                              B")
```
